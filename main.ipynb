```python
from diffusers import StableDiffusionPipeline
# from diffusers.pipelines import dance_diffusion
# from diffusers.utils import make_image_grid
import torch
```

```python
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
PROMPT = 'hyperrealistic portrait of a man as astronaut, portrait, well lit, cyberpunk,'
model_path = "models/Realistic_Vision_V5.1.ckpt"
MODEL_ID = 'stabilityai/stable-diffusion-2-1'

pipe = StableDiffusionPipeline.from_pretrained(MODEL_ID, torch_dtype=torch.float16,).to("mps")

```

```python

result = pipe(PROMPT, num_inference_steps=20, guidance_scale=9).images[0]

result.save('result.png')
```
